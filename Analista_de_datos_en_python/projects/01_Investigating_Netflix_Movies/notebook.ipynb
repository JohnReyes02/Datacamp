#### What was the most frequent *movie* duration in the 1990s? Save an aproximate answer as an integer called `duration` (use 1990 as decades's start year).

```python
# Exploramos los datos que se encuentran en las columnas del dataframe
netflix_df.info()
```

```python
# filtramos el datafrme con el tipo 'Movie'
netflix_movie_df = netflix_df[netflix_df['type'] == 'Movie']
# netflix_movie_df

# Filtramos el dataframe con 'release_year' desde 1990 hasta 1999
netflix_movie_90s_df = netflix_movie_df[(netflix_movie_df['release_year'] >= 1990) & (netflix_movie_df['release_year'] < 2000)]
# netflix_movie_90s_df

# Utilizamos una gráfica de histograma para determinar una aproximacion en cuál es la duración más frecuente.
plt.hist(netflix_movie_90s_df['duration'], alpha=0.7)
plt.title('Distribution of Movie Duration in the 1990s')
plt.xlabel('Duration (minutes)')
plt.ylabel('Number of Movies')
plt.show()
```

```python
# Del histograma anterior se puede deducir que la duración de película que más se repite es aproximadamente 100 minutos
duration = 100
```

#### A movie is considered short if it is less than 90 minutes. Count the number of *short action movies* released in the 1990s and save this integer as `short_movie_count`.

```python
# Filtramos el dataframe de las peliculas de la decada de los 90s, aquellas peliculas con menos de 90 minutos
netflix_movie_90s_short_df = netflix_movie_90s_df[netflix_movie_90s_df['duration'] < 90]

# iniciamos un contador en cero e iteramos en la columna 'genre' incrementando el contador cada vez que encuentre 'Action'
short_movie_count = 0

for label, row in netflix_movie_90s_short_df.iterrows() :
    if row['genre'] == 'Action' :
        short_movie_count += 1

print(short_movie_count)  
```

```python
# Otra forma más optima de hacerlo
short_movie_count2 = (netflix_movie_90s_short_df['genre'] == 'Action').sum()
print(short_movie_count2)
```
